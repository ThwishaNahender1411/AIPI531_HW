```python
## Training environment of Cartpole
env = gym.make('CartPole-v1')

## Evalutation environment of Cartpole
eval_env = gym.make('CartPole-v1')
```

## 1. Vanilla Advantage (gae_lambda = 0.0)

```python
vanilla_model = A2C("MlpPolicy", env, verbose=1, gae_lambda=0, tensorboard_log="./vanilla_CartPole_v1_tensorboard/")
vanilla_model.learn(total_timesteps=100000)
```

```python
eval_env = gym.make('CartPole-v1')
mean_reward_vanilla, std_reward_vanilla = evaluate_policy(vanilla_model, eval_env, n_eval_episodes=100)
print(f"mean_reward:{mean_reward_vanilla:.2f} +/- {std_reward_vanilla:.2f}")
```

## 2. N-Step Advantage (gae_lambda=0.95)

```python
env_1 = gym.make('CartPole-v1')

## Evalutation environment of Cartpole
eval_env_1 = gym.make('CartPole-v1')
```

```python
n_model_1 = A2C("MlpPolicy", env_1, verbose=1, gae_lambda=0.95 , tensorboard_log="./n_step_CartPole_v1_tensorboard/")
n_model_1.learn(total_timesteps=100000)
```

```python
mean_reward_n_1, std_reward_n_1 = evaluate_policy(n_model_1, eval_env_1, n_eval_episodes=100)
print(f"mean_reward:{mean_reward_n_1:.2f} +/- {std_reward_n_1:.2f}")
```

## 3. MC Advantage (gae_lambda=1)

```python
env_2 = gym.make('CartPole-v1')

## Evalutation environment of Cartpole
eval_env_2 = gym.make('CartPole-v1')
```

```python
n_model_2 = A2C("MlpPolicy", env_2, verbose=1, gae_lambda=1, tensorboard_log="./n_step_2_CartPole_v1_tensorboard/")
n_model_2.learn(total_timesteps=100000)
```

```python
mean_reward_n_2, std_reward_n_2 = evaluate_policy(n_model_2, eval_env_2, n_eval_episodes=100)
print(f"mean_reward:{mean_reward_n_2:.2f} +/- {std_reward_n_2:.2f}")
```

## Comparison

We can see that the mean reward got for N-step advantage is inbetween the MC Advantage and Vanilla Advantage

The vanilla advantage is biased with smaller variance since it only considers 1 step
